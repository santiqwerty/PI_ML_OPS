```python
df_games = pd.read_csv('../data/interim/converted_games.csv')
```

```python
print("\nInfo \n", df_games.info())
print("\nTipos Unicos\n")
for column in df_games.columns:
    print(f'{column}: {df_games[column].apply(type).unique()}')
print("\nValoresUnicos \n", df_games.nunique())
print("Head\n", df_games.head())
print("\nTail \n", df_games.tail())
print("\nNulls \n", df_games.isnull().sum())
print("\nDuplicados \n", df_games.duplicated().sum())
print("\nDescribe \n", df_games.describe(include='all'))





```

```python
# Porcentaje de filas con al menos un valor nulo
porcentaje_filas_con_nulos = (df_games.isnull().any(axis=1).sum() / len(df_games)) * 100

# Porcentaje de columnas con al menos un valor nulo
porcentaje_columnas_con_nulos = (df_games.isnull().any(axis=0).sum() / len(df_games.columns)) * 100

# Media, Mediana y Moda de cada columna numérica
# La función describe() de pandas proporciona estadísticas descriptivas incluyendo la media y la mediana
# La función mode() de pandas proporciona la moda
descriptivas = df_games.describe().T[['mean', '50%']]
moda = df_games.mode().iloc[0]

# Combinando la media, mediana y moda en un solo DataFrame para una mejor visualización
estadisticas = pd.DataFrame()
estadisticas['Media'] = descriptivas['mean']
estadisticas['Mediana'] = descriptivas['50%']
estadisticas['Moda'] = moda

print(f'Porcentaje de filas con al menos un valor nulo: {porcentaje_filas_con_nulos:.2f}%')
print(f'Porcentaje de columnas con al menos un valor nulo: {porcentaje_columnas_con_nulos:.2f}%')
print('\nEstadísticas de las columnas numéricas:')
print(estadisticas)
```

```python
# Calculando el porcentaje de valores nulos por columna
missing_values_per_column = df_games.isnull().mean() * 100

# Imprimiendo el porcentaje de valores nulos por columna
print("Porcentaje de valores nulos por columna:\n", missing_values_per_column)

# Calcular la cantidad de valores nulos por fila
null_counts_per_row = df_games.isnull().sum(axis=1)

# Calcular la frecuencia de cada cantidad de valores nulos por fila
null_counts_frequencies = null_counts_per_row.value_counts(normalize=True).sort_index() * 100

# Crear un DataFrame para mostrar los resultados
null_counts_table = pd.DataFrame({
    'Cantidad de valores nulos por fila': null_counts_frequencies.index,
    'Porcentaje de filas': null_counts_frequencies.values
})
print(null_counts_table)
```

```python
# Mantener solo las filas con al menos 13 valores no nulos
df_games_clean = df_games.dropna(thresh=13)

```

```python
# Calculando el porcentaje de valores nulos por columna
missing_values_per_column = df_games_clean.isnull().mean() * 100

# Imprimiendo el porcentaje de valores nulos por columna
print("Porcentaje de valores nulos por columna:\n", missing_values_per_column)

# Calcular la cantidad de valores nulos por fila
null_counts_per_row = df_games_clean.isnull().sum(axis=1)

# Calcular la frecuencia de cada cantidad de valores nulos por fila
null_counts_frequencies = null_counts_per_row.value_counts(normalize=True).sort_index() * 100

# Crear un DataFrame para mostrar los resultados
null_counts_table = pd.DataFrame({
    'Cantidad de valores nulos por fila': null_counts_frequencies.index,
    'Porcentaje de filas': null_counts_frequencies.values
})
print(null_counts_table)

print(df_games_clean.info())
```
